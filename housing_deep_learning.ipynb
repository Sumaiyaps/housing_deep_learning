```python
le = LabelEncoder()
data["mainroad"]=le.fit_transform(data["mainroad"])
data["guestroom"]=le.fit_transform(data["guestroom"])
data["basement"]=le.fit_transform(data["basement"])
data["hotwaterheating"]=le.fit_transform(data["hotwaterheating"])
data["airconditioning"]=le.fit_transform(data["airconditioning"])
data["prefarea"]=le.fit_transform(data["prefarea"])
data["furnishingstatus"]=le.fit_transform(data["furnishingstatus"])




```

```python
data.drop(["basement","hotwaterheating"],axis=1,inplace=True)
```

```python
from sklearn. linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data.info()
```

```python
x=data[["area","bedrooms","bathrooms","stories","mainroad","guestroom","airconditioning","parking","prefarea","furnishingstatus"]]
y=data["price"]
```

```python
data.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
print("r2_score is:",r2_score(ypred,y_test))
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
model=Sequential([
    Dense(1,activation='relu',input_shape=(10,)),
    Dense(1)
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,epochs=200,verbose=1)
```

```python
ypred=model.predict(np.array([[6500,3,2,3,1,0,1,0,1,0]]))
print("predictions for the test set:",ypred)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
ypred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,ypred)
mae=mean_absolute_error(y_test,ypred)
print(f"mean squared error:{mse}")
print(f"mean absolute error:{mae}")
```
